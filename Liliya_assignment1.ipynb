```python
import pandas as pd

# Read the dataset
mlmdata = pd.read_csv('ml1mdata2000.csv')

# Convert timestamp to date
mlmdata['date'] = pd.to_datetime(mlmdata['timestamp'], unit='s', errors='coerce')

# Extract the year
mlmdata['year'] = pd.DatetimeIndex(mlmdata['date']).year

# Get unique years
unique_years = mlmdata['year'].unique()

print("Unique years featured in the dataset:")
print(unique_years)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read the dataset
mlmdata = pd.read_csv('ml1mdata2000.csv')

# Extract the first ten rows
first_10_rows = mlmdata.head(10)

# Create a line plot for movie ratings
plt.figure(figsize=(10, 6))
sns.lineplot(x=first_10_rows.index, y=first_10_rows['rating'],marker="o")
plt.title('Movie Ratings Over the First Ten Rows')
plt.xlabel('Row Index')
plt.ylabel('Rating')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read the dataset
mlmdata = pd.read_csv('ml1mdata2000.csv')

# Select the first one hundred rows
first_100_rows = mlmdata.head(100)

# Create a scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='rating', y='age', data=first_100_rows, alpha=0.5)
plt.title('Movie Ratings vs. Age (First 100 Rows)')
plt.xlabel('Rating')
plt.ylabel('Age')
plt.grid(True)
plt.show()
```
