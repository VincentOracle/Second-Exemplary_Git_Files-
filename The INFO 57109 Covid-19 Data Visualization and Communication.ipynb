## Visualization

```python
#Importing the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
print("Libraries loaded successfully!")
```

```python
# Load datasets
summary_data = pd.read_csv("C:\\Users\\n\Desktop\\worldometer_coronavirus_summary_data.csv")
daily_data = pd.read_csv("C:\\Users\\n\Desktop\\worldometer_coronavirus_daily_data.csv")
print("Dataset loaded successfully")
```

```python
# Task 1: Lineplot Comparing Cumulative Total Cases and Deaths
plt.figure(figsize=(12, 6))
sns.lineplot(data=daily_data, x="date", y="cumulative_total_cases", hue="country")
plt.title("Cumulative Total Cases Over Time")
plt.xlabel("Date")
plt.ylabel("Cumulative Total Cases")
plt.xticks(rotation=45)
plt.legend(loc="upper left")
plt.tight_layout()
plt.show()

```

```python
# Task 2: Lineplot Examining Daily New Cases and Deaths Trends
plt.figure(figsize=(12, 6))
sns.lineplot(data=daily_data, x="date", y="daily_new_cases", hue="country")
plt.title("Daily New Cases Over Time")
plt.xlabel("Date")
plt.ylabel("Daily New Cases")
plt.xticks(rotation=45)
plt.legend(loc="upper left")
plt.tight_layout()
plt.show()

```

```python
# Task 3: Scatterplot  Comparing Testing Rates and Population Adjusted Metrics
plt.figure(figsize=(12, 6))
sns.scatterplot(data=summary_data, x="total_tests_per_1m_population", y="total_cases_per_1m_population", hue="continent", size="population", sizes=(20, 2000), alpha=0.5)
plt.title("Testing Rates vs Total Cases per 1M Population")
plt.xlabel("Total Tests per 1M Population")
plt.ylabel("Total Cases per 1M Population")
plt.xscale("log")
plt.yscale("log")
plt.legend(loc="upper left")
plt.tight_layout()
plt.show()
```

```python

# Task 4: Barplot Comparing Total Deaths Across Continents
plt.figure(figsize=(10, 6))
sns.barplot(data=summary_data, x="continent", y="total_deaths", estimator=sum, ci=None)
plt.title("Total Deaths Across Continents")
plt.xlabel("Continent")
plt.ylabel("Total Deaths")
plt.tight_layout()
plt.show()
```

```python
# Task 5: Active Cases Trend
selected_countries = ["US", "India", "Brazil", "Russia", "UK"]
plt.figure(figsize=(12, 6))
for country in selected_countries:
    country_data = daily_data[daily_data["country"] == country]
    sns.lineplot(data=country_data, x="date", y="active_cases", label=country)
plt.title("Active Cases Trend")
plt.xlabel("Date")
plt.ylabel("Active Cases")
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Task 6: Boxplot for Serious or Critical Cases Distribution
plt.figure(figsize=(12, 6))
sns.boxplot(data=summary_data, x="continent", y="serious_or_critical")
plt.title("Distribution of Serious or Critical Cases Across Continents")
plt.xlabel("Continent")
plt.ylabel("Serious or Critical Cases")
plt.tight_layout()
plt.show()
```

```python

# Task 7: Histogram of Total Confirmed Cases
plt.figure(figsize=(10, 6))
sns.histplot(summary_data["total_confirmed"], bins=20, kde=True)
plt.title("Distribution of Total Confirmed Cases")
plt.xlabel("Total Confirmed Cases")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

```python

# Task 8: Heatmap of Total Cases by Date and Country
# Pivot daily_data to get total cases by date and country
heatmap_data = daily_data.pivot_table(index="date", columns="country", values="cumulative_total_cases", aggfunc="sum")
plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data, cmap="YlGnBu", linewidths=0.5)
plt.title("Total Cases by Date and Country")
plt.xlabel("Country")
plt.ylabel("Date")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
